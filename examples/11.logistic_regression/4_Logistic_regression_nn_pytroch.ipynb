# Logistic Regression as a Neural Network (PyTorch)

This notebook shows a **simple neural network** for binary classification.

- Model: one linear layer (`nn.Linear`)  
- Output: one logit (score)  
- Loss: `BCEWithLogitsLoss` (sigmoid + binary cross entropy together)  
- Optimizer: `SGD` (gradient descent)

Even though this is logistic regression, we build it with standard PyTorch neural-network tools so the workflow matches larger deep-learning models.

## Simple takeaway

This is logistic regression built with neural-network tools:

1. `Linear` layer gives logits,
2. `BCEWithLogitsLoss` computes loss,
3. `SGD` updates weights using gradients.

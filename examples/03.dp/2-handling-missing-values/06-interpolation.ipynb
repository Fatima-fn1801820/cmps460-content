A simple example of missing value imputation using linear interpolation involves estimating a missing value as the midpoint between its surrounding known data points, assuming a straight-line relationship between them.

```python
import pandas as pd
import numpy as np

# Create a DataFrame with a missing value (np.nan)
df = pd.DataFrame({
    'Day': [1, 2, 3, 4, 5],
    'Temperature': [20, 22, np.nan, 26, 28]
})

# Impute the missing value using linear interpolation method
df['Temperature'] = df['Temperature'].interpolate(method='linear')

print(df)
```

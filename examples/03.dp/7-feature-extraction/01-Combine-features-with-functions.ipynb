# Combine features with functions

Creating new features from existing ones using mathematical operations. A practical and common example is calculating the Body Mass Index (BMI) using weight and height.

```python
import pandas as pd
```

```python
# create a dataframe with student data
df = pd.DataFrame({
    'student_id': [1, 2, 3, 4, 5],
    'height_cm': [170, 160, 175, 180, 165],
    'weight_kg': [70, 55, 80, 90, 60]
})

display(df)
```

## Calculate BMI

The formula for BMI is:

$$ BMI = \frac{weight (kg)}{(height (m))^2} $$

We can perform this calculation directly using vectorized operations on the dataframe columns.

```python
# calculate BMI using vectorized operations
# Note: we divide height by 100 to convert cm to meters
df['bmi'] = df['weight_kg'] / (df['height_cm'] / 100) ** 2

display(df)
```

## Create Categories with Custom Functions

Sometimes we want to apply more complex logic to create features. We can use python functions for this. For example, let's categorize the BMI values.

```python
def get_bmi_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal'
    elif 25 <= bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

# apply the function to the bmi column
df['bmi_category'] = df['bmi'].apply(get_bmi_category)

display(df)
```

### One Hot Encoding - sklearn

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# sklearn
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
```

```python
# load titanic dataset

usecols = ["Pclass", "SibSp", "Parch", "Gender", "Embarked", "Cabin", "Survived"]
file_url = "https://raw.githubusercontent.com/cmps460s26/cmps460-content/refs/heads/main/examples/02.eda/datasets/titanic.csv"
data = pd.read_csv(file_url, usecols=usecols)
data["Cabin"] = data["Cabin"].str[0]

data.head()
```

### Split into Train and Test before encoding

Just like imputation, all methods of categorical encoding should be performed over the training set, and then propagated to the test set. 


```python
# let's separate into training and testing set

X_train, X_test, y_train, y_test = train_test_split(
    data.drop("Survived", axis=1),  # predictors
    data["Survived"],  # target
    test_size=0.3,  # percentage of obs in test set
    random_state=0,
)  # seed to ensure reproducibility

X_train.shape, X_test.shape
```

```python
# To start, we fillna manually. Later on
# we add this step to a pipeline

X_train.fillna("Missing", inplace=True)
X_test.fillna("Missing", inplace=True)
```

```python
# set up encoder
# sparse_output=False -> All 0/1 columns are stored explicitly
encoder = OneHotEncoder(sparse_output=False)
encoder.set_output(transform="pandas")
```

```python
# fit the encoder (finds categories)

encoder.fit(X_train)
```

```python
# we observe the learned categories
encoder.categories_
```

```python
# transform the data sets

X_train_enc = encoder.transform(X_train)
X_test_enc = encoder.transform(X_test)

X_train_enc.head()
```

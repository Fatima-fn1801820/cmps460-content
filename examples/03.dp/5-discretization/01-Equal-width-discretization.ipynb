## Discretization

Discretization is the process of transforming continuous variables into discrete variables by creating a set of contiguous intervals that span the range of the variable's values. Discretization is also called **binning**, where bin is an alternative name for interval.

Equal‑Width Discretization, commonly used in housing price prediction, customer income grouping, or risk scoring.


## Equal width discretization

Equal width discretization divides the scope of possible values into N bins of the same width. The width is determined by the range of values in the variable and the number of bins we wish to use to divide the variable:

There is no rule of thumb to define N; that is something to determine experimentally.

```python
import pandas as pd
income = pd.Series([25, 32, 45, 50, 61, 75, 82, 95, 120, 150])
```

We want to group these into 4 equal‑width income categories:
Low income
Lower‑middle
Upper‑middle
High income

```python
bins = pd.cut(
    income,
    bins=4,
    labels=["Low", "Lower-Mid", "Upper-Mid", "High"]
)

df = pd.DataFrame({"Income(k$)": income, "Category": bins})
display(df)
```

🔍 How the bins were formed
Minimum = 25
Maximum = 150
Range = 125
With 4 bins:

Bin width = 125 / 4 = 31.25
So the equal‑width bins are:

25.00 → 56.25 → "Low"
56.25 → 87.50 → "Lower‑Mid"
87.50 → 118.75 → "Upper‑Mid"
118.75 → 150.00 → "High"

Equal‑width discretization is simple and helps:
✔ reduce noise
✔ simplify modeling
✔ prepare data for Naive Bayes or other algorithms needing categorical input

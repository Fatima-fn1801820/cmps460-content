## 🎲 Efficient Tuning: RandomizedSearchCV
When your hyperparameter space is massive, exhaustive Grid Search becomes computationally impossible. Random Search is the modern solution.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

# Define the parameter distributions
param_dist = {
    'max_depth': randint(3, 50), # Random integer between 3 and 50
    'min_samples_leaf': randint(1, 20),
    'max_features': ['sqrt', 'log2', None]
}

# Initialize RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=DecisionTreeClassifier(random_state=42),
    param_distributions=param_dist,
    n_iter=20,  # Number of random combinations to try
    cv=5,
    scoring='accuracy',
    random_state=42,
    n_jobs=-1
)

# Fit the random search
random_search.fit(X_train, y_train)

print(f"Best Cross-Validation Accuracy: {random_search.best_score_:.4f}")
print(f"Optimal Hyperparameters: {random_search.best_params_}")
```

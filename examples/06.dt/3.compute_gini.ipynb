### Computing Gini Impurity

The **Gini Impurity** measures how impure (mixed) a node is. 
- **Lower Gini (closer to 0)**: The node is more *pure* (most elements belong to one class).
- **Higher Gini**: The node is more *mixed*.

**Formula**:
$$Gini = 1 - \sum_{i=1}^{C} (p_i)^2$$
where $p_i$ is the probability of an item belonging to class $i$.

```python
from collections import Counter

def gini_impurity(labels):
    """
    Compute the Gini Impurity for a list of labels.
    """
    if len(labels) == 0:
        return 0.0
    
    # 1. Count frequencies of each class
    counts = Counter(labels)
    total = len(labels)
    
    # 2. Compute probabilities (p_i) and Gini Impurity: 1 - sum(p_i^2)
    probabilities = (count / total for count in counts.values())
    return 1 - sum(p ** 2 for p in probabilities)

# Example 1: A perfectly mixed dataset of 3 classes
mixed_classes = ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C']
print(f"Gini Impurity (Mixed): {gini_impurity(mixed_classes):.2f}")

# Example 2: A pure dataset
pure_classes = ['A', 'A', 'A', 'A']
print(f"Gini Impurity (Pure): {gini_impurity(pure_classes):.2f}")
```

### Evaluating a Split (Weighted Gini Impurity)

When evaluating a potential split in a decision tree, we must measure the impurity of the resulting child nodes. The overall impurity of the split is the **weighted average** of the Gini Impurity of each child node.

**Formula**:
$$Gini_{split} = \sum_{j=1}^{K} \frac{n_j}{N} Gini(Node_j)$$
where:
- $K$ is the number of child nodes (e.g., 2 for a binary split).
- $n_j$ is the number of samples in child node $j$.
- $N$ is the total number of samples before the split.
- $Gini(Node_j)$ is the Gini Impurity of child node $j$.

```python
# Real-world Example: Predicting if a customer 'Bought Product' (Yes/No) based on 'Gender' split
# Target variable: 'Yes' (Bought), 'No' (Did not buy)

female_node = ['Yes', 'Yes', 'Yes', 'Yes', 'No'] # 4 bought, 1 didn't buy
male_node = ['No', 'No', 'No', 'Yes', 'Yes']     # 2 bought, 3 didn't buy

gini_female = gini_impurity(female_node)
gini_male = gini_impurity(male_node)

print(f"Gini Impurity (Female Node): {gini_female:.2f}")
print(f"Gini Impurity (Male Node):   {gini_male:.2f}")

# Compute the Weighted Average Gini Impurity for the split
total_samples = len(female_node) + len(male_node)

weight_female = len(female_node) / total_samples
weight_male = len(male_node) / total_samples

gini_split = (weight_female * gini_female) + (weight_male * gini_male)
print(f"\nWeighted Average Gini Impurity for 'Gender' split: {gini_split:.3f}")
```

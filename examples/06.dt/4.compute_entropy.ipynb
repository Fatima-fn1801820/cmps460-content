### Computing Entropy

**Entropy** is another metric used to measure the impurity or uncertainty of a node. 
- **Lower Entropy (closer to 0)**: The node is more *pure* (most elements belong to one class).
- **Higher Entropy**: The node is more *mixed*.

**Formula**:
$$Entropy = - \sum_{i=1}^{C} p_i \log_2(p_i)$$
where $p_i$ is the probability of an item belonging to class $i$.

```python
import math
from collections import Counter

def compute_entropy(labels):
    """
    Compute the Entropy for a list of labels.
    """
    if len(labels) == 0:
        return 0.0
    
    # 1. Count frequencies of each class
    counts = Counter(labels)
    total = len(labels)
    
    # 2. Compute probabilities (p_i) and Entropy: -sum(p_i * log2(p_i))
    probabilities = (count / total for count in counts.values())
    return -sum(p * math.log2(p) for p in probabilities if p > 0)

# Example 1: A perfectly mixed dataset of 3 classes
mixed_classes = ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C']
print(f"Entropy (Mixed): {compute_entropy(mixed_classes):.2f}")

# Example 2: A pure dataset
pure_classes = ['A', 'A', 'A', 'A']
print(f"Entropy (Pure): {compute_entropy(pure_classes):.2f}")
```

### Evaluating a Split (Weighted Entropy)

Similar to Gini Impurity, we evaluate a potential split by calculating the **weighted average** of the Entropy of each child node.

**Formula**:
$$Entropy_{split} = \sum_{j=1}^{K} \frac{n_j}{N} Entropy(Node_j)$$
where:
- $K$ is the number of child nodes (e.g., 2 for a binary split).
- $n_j$ is the number of samples in child node $j$.
- $N$ is the total number of samples before the split.
- $Entropy(Node_j)$ is the Entropy of child node $j$.

```python
# Real-world Example: Predicting if a customer 'Bought Product' (Yes/No) based on 'Gender' split
# Target variable: 'Yes' (Bought), 'No' (Did not buy)

female_node = ['Yes', 'Yes', 'Yes', 'Yes', 'No'] # 4 bought, 1 didn't buy
male_node = ['No', 'No', 'No', 'Yes', 'Yes']     # 2 bought, 3 didn't buy

entropy_female = compute_entropy(female_node)
entropy_male = compute_entropy(male_node)

print(f"Entropy (Female Node): {entropy_female:.2f}")
print(f"Entropy (Male Node):   {entropy_male:.2f}")

# Compute the Weighted Average Entropy for the split
total_samples = len(female_node) + len(male_node)

weight_female = len(female_node) / total_samples
weight_male = len(male_node) / total_samples

entropy_split = (weight_female * entropy_female) + (weight_male * entropy_male)
print(f"\nWeighted Average Entropy for 'Gender' split: {entropy_split:.3f}")
```

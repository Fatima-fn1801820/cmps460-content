# Gradient Descent on $x^2$

We want to minimize:

$$f(x)=x^2$$

The minimum value is at **x = 0**.

Gradient descent helps us move toward 0 step by step.


```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1) Function and Gradient

- Function: $f(x)=x^2$
- Gradient (slope): $\frac{df}{dx}=2x$

If slope is positive, move left.
If slope is negative, move right.


```python
def f(x):
    return x**2


def grad_f(x):
    return 2 * x


x0 = 3.0
print("Start x:", x0)
print("f(x):", f(x0))
print("gradient:", grad_f(x0))
```

## 2) Update Rule

Gradient descent update:

$$x_{new}=x-\alpha\cdot 2x$$

- $\alpha$ is the learning rate (step size).
- We repeat this many times.


```python
alpha = 0.1
x = 3.0
steps = 15

history = [x]

for _ in range(steps):
    x = x - alpha * grad_f(x)
    history.append(x)
```

## 3) Values Over Time

You should see x getting closer to 0.


```python
for i, x_val in enumerate(history[:8]):
    print(f"step {i:2d}: x = {x_val:.6f}, f(x) = {f(x_val):.6f}")
```

## 4) Visualize the Path

Dots show how gradient descent moves toward the minimum.


```python
x_curve = np.linspace(-3.5, 3.5, 400)
y_curve = f(x_curve)

y_hist = [f(v) for v in history]

plt.figure(figsize=(7, 4))
plt.plot(x_curve, y_curve, label="f(x)=x^2")
plt.scatter(history, y_hist, color="red", s=25, label="gradient descent steps")
plt.title("Gradient Descent on x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

## Key Takeaway

- Gradient descent uses slope to decide direction.
- For $f(x)=x^2$, it keeps moving toward **x = 0**.
- Smaller function value means we are closer to the minimum.

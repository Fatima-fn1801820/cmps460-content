# Polynomial Regression from Scratch

We will predict **house price** from **house size** using a polynomial model:

\[
\hat{y} = w_0 + w_1x + w_2x^2
\]

This lets the model capture gentle curvature that plain linear regression cannot.

```python
import numpy as np
import matplotlib.pyplot as plt

# x: house size in 1000 sq ft, y: price in $100k
x = np.array([0.9, 1.1, 1.3, 1.6, 1.9, 2.2, 2.6, 3.0, 3.3])
y = np.array([1.2, 1.35, 1.55, 1.95, 2.3, 2.75, 3.4, 4.15, 4.9])

plt.scatter(x, y, color='royalblue', label='Training data')
plt.xlabel('Size (1000 sq ft)')
plt.ylabel('Price ($100k)')
plt.title('House Size vs Price')
plt.grid(alpha=0.25)
plt.legend()
plt.show()
```

## 1) Build polynomial features

A polynomial model is still linear in parameters. We just expand each input:

\[
x \rightarrow [1, x, x^2]
\]

Then we learn one weight per column.

```python
def poly_features(x):
    return np.column_stack([np.ones_like(x), x, x**2])

X = poly_features(x)
print('X shape:', X.shape)
print('First row [1, x, x^2]:', X[0])
```

## 2) Cost and gradients

We use Mean Squared Error:

\[
J(w)=\frac{1}{2m}\sum_{i=1}^{m}(\hat{y}^{(i)}-y^{(i)})^2
\]

Gradient descent repeatedly updates weights in the opposite gradient direction.

```python
def compute_cost(X, y, w):
    m = len(y)
    y_hat = X @ w
    return (1 / (2 * m)) * np.sum((y_hat - y) ** 2)


def gradient_descent(X, y, w_init, alpha=0.1, iters=5000):
    m = len(y)
    w = w_init.astype(float).copy()
    history = []

    for _ in range(iters):
        y_hat = X @ w
        grad = (1 / m) * (X.T @ (y_hat - y))
        w -= alpha * grad
        history.append(compute_cost(X, y, w))

    return w, history
```

## 3) Train and visualize

We initialize weights at zero, run gradient descent, and draw the learned curve on top of the data.

```python
w0 = np.zeros(X.shape[1])
w, history = gradient_descent(X, y, w0, alpha=0.1, iters=5000)

print('Learned weights [w0, w1, w2]:', np.round(w, 4))
print('Final cost:', round(history[-1], 6))

# Smooth curve for plotting
x_line = np.linspace(x.min(), x.max(), 200)
X_line = poly_features(x_line)
y_line = X_line @ w

plt.figure(figsize=(7, 4))
plt.scatter(x, y, color='royalblue', label='Training data')
plt.plot(x_line, y_line, color='crimson', linewidth=2, label='Polynomial fit (degree 2)')
plt.xlabel('Size (1000 sq ft)')
plt.ylabel('Price ($100k)')
plt.title('Polynomial Regression Fit')
plt.grid(alpha=0.25)
plt.legend()
plt.show()

plt.figure(figsize=(7, 3.5))
plt.plot(history, color='darkgreen')
plt.xlabel('Iteration')
plt.ylabel('Cost J(w)')
plt.title('Cost Decreases During Gradient Descent')
plt.grid(alpha=0.25)
plt.show()
```

## Quick recap

- We created polynomial features manually (`1, x, x^2`).
- We implemented cost and gradient descent from scratch.
- The fitted curve captures the nonlinear trend in house prices.

Try changing the degree (for example add `x^3`) and compare the new fit.

```python
# Gradient Descent for Linear Regression (multiple variables)
# Model: y_hat = X @ w + b
# Loss:  MSE = mean((y_hat - y)^2)
import numpy as np

np.set_printoptions(
    suppress=True,
    formatter={"float_kind": "{:0.6f}".format},
)
```

## Gradient Descent (Multiple Variables)

We learn two types of parameters:
- `w`: one weight per feature
- `b`: one bias/intercept term

At each epoch:
1. Predict: `y_hat = X @ w + b`
2. Error: `error = y_hat - y`
3. Gradients:
   - `dw = (1/m) * (X.T @ error)`
   - `db = (1/m) * sum(error)`
4. Update:
   - `w = w - lr * dw`
   - `b = b - lr * db`

This keeps the implementation vectorized, short, and easy to debug.

```python
# Tiny vectorization check (optional intuition)
X_demo = np.array([[1, 2], [3, 4]])
w_demo = np.array([0.5, 1.0])
b_demo = 0.2

print("X_demo shape:", X_demo.shape)
print("w_demo shape:", w_demo.shape)
print("prediction:", X_demo @ w_demo + b_demo)
print("w_demo transpose:", w_demo.T)
```

```python
def compute_mse(X, y, w, b):
    """Mean Squared Error for y_hat = X @ w + b."""
    # @ is dot product operator
    y_hat = X @ w + b
    return np.mean((y_hat - y) ** 2)


def gradient_step(X, y, w, b, lr):
    """One gradient descent step (vectorized)."""
    m = X.shape[0]
    error = (X @ w + b) - y

    dw = (1 / m) * (X.T @ error)
    db = (1 / m) * np.sum(error)

    w = w - lr * dw
    b = b - lr * db
    return w, b
```

```python
def gradient_descent(X, y, lr=0.01, epochs=1000, debug_every=100):
    """Train linear regression with explicit weights (w) and bias (b)."""
    m, n = X.shape

    w = np.zeros(n)
    b = 0.0

    for epoch in range(epochs):
        w, b = gradient_step(X, y, w, b, lr)

        # Simple educational debugging during learning
        if epoch % debug_every == 0 or epoch == epochs - 1:
            mse = compute_mse(X, y, w, b)
            print(f"epoch={epoch:4d} | mse={mse:10.6f} | w={w} | b={b:0.6f}")

    return w, b
```

```python
def predict(X, w, b):
    return X @ w + b
```

```python
# Usage example
X = np.array([
    [1, 2],
    [2, 3],
    [3, 4],
    [4, 5],
], dtype=float)

y = np.array([6, 10, 12, 23], dtype=float)

print("X shape:", X.shape, "| y shape:", y.shape)

w, b = gradient_descent(
    X, y,
    lr=0.001,
    epochs=1000,
    debug_every=100,
)

print("\nLearned parameters")
print("w:", w)
print("b:", b)

X_test = np.array([[5, 6], [6, 7]], dtype=float)
y_pred = predict(X_test, w, b)
print("\nX_test:")
print(X_test)
print("Predictions:", y_pred)
```

# Cost Function

The **cost function** tells us how wrong our line is.

For linear regression:

$$f_{w,b}(x)=wx+b$$
$$J(w,b)=\frac{1}{2m}\sum_{i=1}^{m}(f_{w,b}(x^{(i)})-y^{(i)})^2$$

- If predictions are close to actual values, cost is small.
- If predictions are far, cost is large.
- Squaring makes all errors positive and penalizes large errors more.


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Tiny Dataset

Assume house size (`x`) is in thousands of square feet and price (`y`) is in thousands of dollars.


```python
x_train = np.array([1.0, 2.0, 3.0])
y_train = np.array([300.0, 500.0, 700.0])

x_train, y_train
```

## Cost Function in Code

This function computes average squared error (with the common $\frac{1}{2m}$ scaling).


```python
def compute_cost(x, y, w, b):
    m = len(x)
    predictions = w * x + b
    errors = predictions - y
    return (1 / (2 * m)) * np.sum(errors ** 2)
```

## Quick Check

Compare a good line vs a bad line.


```python
good_w, good_b = 200, 100
bad_w, bad_b = 120, 150

print("Good line cost:", compute_cost(x_train, y_train, good_w, good_b))
print("Bad line cost :", compute_cost(x_train, y_train, bad_w, bad_b))
```

## Visualization of Cost Function

Keep `b=100` fixed and vary `w`. The minimum point is the best slope for this `b`.


```python
b_fixed = 100
w_values = np.linspace(50, 300, 200)
cost_values = [compute_cost(x_train, y_train, w, b_fixed) for w in w_values]

best_index = int(np.argmin(cost_values))
best_w = w_values[best_index]
best_cost = cost_values[best_index]

plt.figure(figsize=(7, 4))
plt.plot(w_values, cost_values)
plt.scatter([best_w], [best_cost], color="red", label=f"min at w={best_w:.1f}")
plt.title("Cost vs w (b fixed at 100)")
plt.xlabel("w")
plt.ylabel("J(w, b=100)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

### Key Takeaway

Training linear regression means finding `w` and `b` that **minimize cost**.
